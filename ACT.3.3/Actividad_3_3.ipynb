```python
#Para rellenar HOST_LOCATION ocuparÃ© un forward fill ya que en su mayorÃ­a los datos capturados en la base de datos
#original provienen de New Orleans, LA.

#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["host_location"] =df["host_location"].fillna(method="ffill")
df
```

```python
#A partir de este momento, me di cuenta que hay cuatro valores con exactamente la misma cantidad de datos faltantes, y analizando la base de datos
#original, el tipo de datos faltantes es parecido. Por lo tanto, estarÃ© rellenando con 0 para no modificar el comportamiento de la informaciÃ³n og.

df["review_scores_rating"]= df["review_scores_rating"].fillna("0")
df["reviews_per_month"]= df["reviews_per_month"].fillna("0")
df["review_scores_cleanliness"]= df["review_scores_cleanliness"].fillna("0")
df["review_scores_communication"]= df["review_scores_communication"].fillna("0")
```

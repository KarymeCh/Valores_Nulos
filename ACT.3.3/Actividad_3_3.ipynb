```python
#Para rellenar HOST_LOCATION ocuparé un forward fill ya que en su mayoría los datos capturados en la base de datos
#original provienen de New Orleans, LA.

#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["host_location"] =df["host_location"].fillna(method="ffill")
df
```

```python
df["host_response_rate"].value_counts().head(5)
```

```python
#Para host_response_rate ocuparé la moda, ya que los datos faltantes representan el 22%, existe una clara moda con
#el valor de 100% y en la lista completa hay dispersión, por lo cual usar la media no sería tan óptimo.

#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
df["host_response_rate"]= df["host_response_rate"].fillna("100%")
df
```

```python
#A partir de este momento, me di cuenta que hay cuatro valores con exactamente la misma cantidad de datos faltantes, y analizando la base de datos
#original, el tipo de datos faltantes es parecido. Por lo tanto, estaré rellenando con 0 para no modificar el comportamiento de la información og.

df["review_scores_rating"]= df["review_scores_rating"].fillna("0")
df["reviews_per_month"]= df["reviews_per_month"].fillna("0")
df["review_scores_cleanliness"]= df["review_scores_cleanliness"].fillna("0")
df["review_scores_communication"]= df["review_scores_communication"].fillna("0")
```

```python
#para este valor ocuparé la moda
df["beds"]= df["beds"].fillna("2")
```

```python
df["price"]= df["price"].fillna("0")
```

```python
#para el caso del precio, decidí dejarla en valores de 0 ya que existe mucha dispersión, por lo tanto no habría información verosímil si lo reempazara
#por alguna medida de tendencia central
#CONOCER EL PORCENTAJE DE MISSING VALUES POR COLUMNA
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(6)
```

```python
df["bathrooms"].describe()
```

```python
#Para la parte de baños, decidí colocar 1, ya que mínimo deben tener un baño, además de que la moda es ese valor
df["bathrooms"]= df["bathrooms"].fillna("1")
```

```python
df["host_is_superhost"].value_counts()
```

```python
#Apartir de aquí imputaré con medidas de tendencia central ya que el porcentaje faltante es menor al 5%
df["host_is_superhost"]= df["host_is_superhost"].fillna("f")
```

```python
df["has_availability"].describe()
df["has_availability"]= df["has_availability"].fillna("t")
```

```python
df["bedrooms"]= df["bedrooms"].fillna("1")
```

```python
df["bathrooms_text"]= df["bathrooms_text"].fillna("1 bath")
```
